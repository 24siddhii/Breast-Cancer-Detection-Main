```python
X = new_data.values[:, 1:]
print(X)
```

```python
r1= new_data['diagnosis'].replace({'B': 0, 'M': 1})
print(r1)

```

```python
X_train,X_test,y_train, y_test=train_test_split(X, r1, test_size=0.2, random_state=100)
```

```python
scaler=StandardScaler()
X1_train_std=scaler.fit_transform(X_train)
X1_test_std=scaler.transform(X_test)
```

```python
#SVM classifier
classifier1 = SVC(kernel='linear')

classifier1.fit(X_train, y_train)
y1_pred = classifier1.predict(X_test)
print("Accuracy for SVM:", accuracy_score(y_test, y1_pred))
print()
print(classification_report(y_test, y1_pred))

```

```python
#KNN classifier
classifier2 = KNeighborsClassifier(n_neighbors=3 )

# model training
classifier2.fit(X_train,y_train)

# testing the model
y2_pred= classifier2.predict(X_test)
# printing accuracy
print("Accuracy for KNN:", accuracy_score(y_test,y2_pred))
print()
print(classification_report(y_test, y2_pred))

```

```python
#Naive Bayes classifier
classifier3 = GaussianNB()

# training the model
classifier3.fit(X_train, y_train)

# testing the model
y3_pred = classifier3.predict(X_test)
print("Accuracy for Naive Bayes: ", accuracy_score(y_test,y3_pred))
print()
print(classification_report(y_test, y3_pred))
```

```python
#Random Forest classifier
classifier4 = RandomForestClassifier()

# fit the model
classifier4.fit(X_train, y_train)

y4_pred = classifier4.predict(X_test)
print("Accuracy score for Random forest:", accuracy_score(y_test,y4_pred))
print()
print(classification_report(y_test, y4_pred))
```

```python
voting_clf = VotingClassifier(estimators = [('svm',classifier1 ), ('knn', classifier2 ),
                                            ('nb', classifier3),('rf', classifier4)], voting = 'hard')
voting_clf.fit(X_train, y_train)
```

```python
predictions = voting_clf.predict(X_test)
print(classification_report(y_test,predictions))
print(accuracy_score(y_test,predictions))
```
